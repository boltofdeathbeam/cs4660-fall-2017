<h1>Convert the dataframe to Numpy Array</h1>

```python
final_data = df.values
final_data
```

<h1>Divide data into training set.</h1>

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(final_data[:,:15], final_data[:,15], test_size=0.33, random_state=0)
X_test
```

<h1>Import SKlearn RandomForestClassifier... and use it</h1>

```python
from sklearn.ensemble import RandomForestClassifier 
from sklearn import metrics

forest = RandomForestClassifier(n_estimators = 100)
forest.fit(X_train, y_train)
output = forest.predict(X_test)
```

<h1>Return the accuracy on the test.</h1>

```python
metrics.accuracy_score(y_test, output)
```

```python
from sklearn.metrics import classification_report
target_names = ['accepted', 'declined']
print classification_report(y_test, output, target_names=target_names)
```
